```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Load dataset
df = pd.read_csv('customer_data_records.csv')
print(df.head())
```

```python
# Select features for clustering
y = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Apply K-means clustering using 2 clusters
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans.fit(y)
```

```python
# Add cluster labels to our dataset
df['Cluster'] = kmeans.labels_
print(df.head())
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 5))
sns.scatterplot(data = df, x='Annual Income (k$)', y='Spending Score (1-100)', hue = 'Cluster')
plt.title('Customer clusters based on annual income and spennding scores')
plt.xlabel('Annual income ($k)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
